```python
import math
```

```python
def bigram_model(w1, w2):
    '''
    Return log probibility of w2 given w1.
    P(w1 -> w2) = cnt(w1 -> w2)/cnt(w2)
    '''
    bigram = (w1, w2)
    try:
        prob = bigram_cnt[bigram]/unigram_cnt[w1]
    except KeyError:
        return float("-inf")
    return math.log10(prob)
```

```python
bigram_model('the', 'fulton')
```
